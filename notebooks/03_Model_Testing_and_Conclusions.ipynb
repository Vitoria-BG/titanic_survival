Agora, importamos o modelo (SVM + Random Forest) e fazemos as análises com x_test. Para submissão no Kaggle, criamos um submission_titanic com PassengerId e o resultado das previsões.

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
import joblib

# Carregando o modelo que será utilizado
model_path = '../models/ensemble_titanic.joblib'
artefacts = joblib.load(model_path)

model = artefacts['modelo_final']
features_train = artefacts['features']

# Carregando as bases de dados
x_test = pd.read_csv('../data/x_test.csv')
ids_passengers = pd.read_csv('../data/ids_passengers.csv')

# Garantir o mesmo alinhamento de colunas
x_test = x_test[features_train]

# Previsoes com o x_test
prediction = model.predict(x_test)

# Criar o DataFrame de Resultados
df_final = pd.DataFrame({
    'PassengerId': ids_passengers.iloc[:, 0].values,
    'Survived': prediction
})

# Salvar o arquivo final para upload
df_final.to_csv('../data/submission_titanic.csv', index=False)
```
